```python
data_n = pd.concat([data.drop(['시스템구분'],axis=1),pd.get_dummies(data['시스템구분'])], axis=1)
data_n = pd.concat([data_n.drop(['업무그룹','업무그룹CD','업무'],axis=1),pd.get_dummies(data['업무그룹'])], axis=1)
data_n = pd.concat([data_n.drop(['프로그램종류'],axis=1),pd.get_dummies(data['프로그램종류'])], axis=1)
data_n = pd.concat([data_n.drop(['디렉토리','디렉토리CD'],axis=1),pd.get_dummies(data['디렉토리CD'])], axis=1)
data_n = pd.concat([data_n.drop(['프로그램명','프로그램설명','담당파트','담당파트CD', '담당자행번'],axis=1),pd.get_dummies(data['담당파트'])], axis=1)
```

```python
from datetime import datetime

def timediff(t):
    date_format = "%Y%m%d"
    a = datetime.strptime(str(t[0]), date_format)
    b = datetime.strptime(str(t[1]), date_format)
    delta = b - a
    return delta.days
timediff = data[['신규등록일','최종변경일']].apply(timediff,axis=1)
timediff = pd.Series(timediff,name='timediff')
data_n=pd.concat([data_n,timediff],axis=1)
```

```python
data_n = pd.concat([data_n.drop(['신규등록일','최종변경일','폐기일','프로그램상태CD'],axis=1),pd.get_dummies(data['프로그램상태'])], axis=1)
```

```python
data_n = data_n.drop(['프로그램상태'],axis=1)
```

```python
data_n.head()
```

```python
data_nn = data_n.groupby(data_n['담당자']).sum()
data_nn
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
fitted = mm.fit(data_nn)
out = mm.transform(data_nn)
out = pd.DataFrame(out, columns=data_nn.columns, index=list(data_nn.index.values))
out.head()
```

```python
data_nn = out.transpose()
data_nn
```

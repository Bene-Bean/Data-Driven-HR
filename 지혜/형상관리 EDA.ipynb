```python
from datetime import datetime

def timediff(t):
    date_format = "%Y%m%d"
    a = datetime.strptime(str(t[0]), date_format)
    b = datetime.strptime(str(t[1]), date_format)
    delta = b - a
    return delta.days

```

```python
timediff = data[['신규등록일','최종변경일']].apply(timediff,axis=1)
timediff
```

```python
timediff = pd.Series(timediff,name='timediff')
```

```python
data=pd.concat([data,timediff],axis=1)
data.head()
```

```python
plt.figure(figsize=(20,5))
sns.distplot(timediff)
```

```python
pd.get_dummies(data['업무'])
```

```python
pd.get_dummies(data['프로그램종류'])
```

```python
d1=pd.get_dummies(data['프로그램종류'])
d2=pd.get_dummies(data['업무'])
cor_mat=pd.concat([d2,d1], axis=1, keys=['프로그램종류','업무']).corr().loc['프로그램종류','업무']
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(cor_mat,cmap='coolwarm',annot=True)
plt.show()
```

```python
d1=pd.get_dummies(data['프로그램상태'])
d2=pd.get_dummies(data['timediff'])
cor_mat=pd.concat([d2,d1], axis=1, keys=['프로그램상태','timediff']).corr().loc['프로그램상태','timediff']
plt.figure(figsize=(30,30))
sns.heatmap(cor_mat,cmap='coolwarm',annot=True)
plt.show()
```

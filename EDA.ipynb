
# Join Data
## Create Jikwon Index

```python
jikwon.shape
```

- 24 Features, 13973 Entries

-> ICT 그룹 직원을 메인 인덱스로 사용하기 위해 점번호를 기준으로 필터링
- 61(ICT운영부)
- 62(정보개발부)
- 63(금융개발부)
- 64(글로벌개발부)
- 66(기관개발부)
- 69(디지털개발부)
- 507(ICT기획부)

```python
main_data = jikwon[(jikwon['JEOM_NO'] == 61) | (jikwon['JEOM_NO'] == 62) | (jikwon['JEOM_NO'] ==  63) |
                   (jikwon['JEOM_NO'] == 64) | (jikwon['JEOM_NO'] == 66) | (jikwon['JEOM_NO'] == 69) | (jikwon['JEOM_NO'] ==  507)]
```

```python
main_data.shape
```

402명으로 직원 수를 줄였다

이를 기반으로 모든 데이터를 JOIN 할 것이다

## Merge Data
### Wrangling Program Data
데이터 조인에 앞서 프로그램 변경이력의 경우 2개의 dataframe으로 나뉘어져 있기 때문에 이를 합치도록 한다

```python
program = pd.concat([program_change_1, program_change_2], axis=0)
```

```python
print(program_change_1.shape)
print(program_change_2.shape)
print(program.shape)
```

또한, 직원 데이터의 직원번호는 **int형**인 반면, 프로그램 데이터의 직원번호는 'BK' + 행번 형태의 **object형**임을 이전에 확인했다

```python
jikwon['JIKWON_NO']
```

```python
program['COL07']
```

프로그램 데이터의 직원번호를 직원데이터와 형태가 일치하도록 수정한다

그 전에, 모든 직원번호가 **BK + 행번** 형태인지 확인하자

```python
program[program['COL07'].str[:2] != 'BK'].COL07
```

이렇듯, BK로 시작하지 않는, 다시 말해 당행 직원이 아닌 데이터가 있다

해당 데이터는 현재 단계에서 필요하지 않기 때문에 제외하도록 한다

```python
print(program.shape, end="->")
program = program[program['COL07'].str[:2] == 'BK']
print(program.shape)
```

이제 다시 직원데이터와 형태가 일치하도록 변경한다

```python
program['COL07'] = program['COL07'].str[2:].astype(int)
```

```python
program.head()
```
